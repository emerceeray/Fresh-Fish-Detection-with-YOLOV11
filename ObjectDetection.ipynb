```python
from ultralytics import YOLO


model = YOLO('yolo11n.pt')
```

```python
model.train(data='/content/dataset/data.yaml', epochs=30, imgsz=768, batch=32)
```

```python
# Eğitilen modeli yükleyin
model = YOLO('/content/runs/detect/train/weights/best.pt')  # En iyi modeli yükleyin

# Test veri kümesinde doğrulama yapın
results = model.val(data='/content/dataset/data.yaml')

# Sonuçları görüntüleyin
print(results)
```

```python
results = model('/content/deneme.jpg', save=True, save_dir='/content/predictions')
```
